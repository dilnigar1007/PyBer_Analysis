```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis, marker="*", linewidth=2, label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="d", color="Green", linewidth=2, label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0, 45)
ax.legend()
ax.grid()
```

```python
plt.bar(x_axis, y_axis, color="green", label="Boston")
plt.xlabel ("Date")
plt.ylabel ("Fare ($)")
plt.title ("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis, y_axis)
```

```python
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, c="green", marker="x", s=100)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("AVerage Fare ($)")
plt.title("Average Fare vs. Total Rides")
plt.show()
```

```python
plt.plot(y_axis,x_axis, "o", c="red", label="Chicago")
plt.xlabel("Fare($)")
plt.ylabel ("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.legend()
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter (x_axis, y_axis, s=y_axis_larger)
plt.scatter (x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
fg, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i*5 for i in y_axis], c="skyblue", edgecolor="black", linewidth=2, label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,50)
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%1.1f%%')
plt.show()
```
